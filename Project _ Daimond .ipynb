__scale your input data__

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(df)
```

**train_test_split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

**Create regression models**

1 Linear Regreesiom

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

```python
# Sum of Square Error
sse=sum((y_pred1-y_test)**2)
print('Sum of Squared Error',sse)

# mean square error
mse=sum((y_pred1-y_test)**2)/len(y_test)
print('mean of Squared Error',mse)

# root mean square error

rmse=np.sqrt(sum((y_pred1-y_test)**2)/len(y_test))
print(' Root mean of Squared Error',rmse)
```
